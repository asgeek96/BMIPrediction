## Project Overview
This project performs **Exploratory Data Analysis (EDA)** on a dataset of 500 individuals.

| Index | Status |
|-------|--------|
| 0 | Extremely Weak |
| 1 | Weak |
| 2 | Normal |
| 3 | Overweight |
| 4 | Obesity |
| 5 | Extreme Obesity |

## Tools Used
- **Pandas** — data loading and manipulation
- **Matplotlib & Seaborn** — data visualization
- **scikit-learn** — preprocessing

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 2. Load the Dataset

We load the dataset from a CSV file containing 500 records with Gender, Height (m), Weight (kg), BMI, and BMI Index.

```python
# Load the dataset
df = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')

# Rename columns for clarity
df.columns = ['Gender', 'Height', 'Weight', 'Index']

# Calculate BMI: weight (kg) / height (m)^2
df['BMI'] = (df['Weight'] / (df['Height'] / 100) ** 2).round(4)

print(f"Dataset shape: {df.shape}")
df.head(10)
```

## 3. Data Overview

Let's check for missing values and understand the basic statistics of the dataset.

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

print("\nBasic statistics:")
df.describe()
```

## 4. Gender Distribution

Let's see how many males and females are in the dataset.

```python
# Count and percentage breakdown by gender
total = len(df)
gender_counts = df['Gender'].value_counts()

print(f"Total people: {total}")
for gender, count in gender_counts.items():
    print(f"  {gender}: {count} ({count/total*100:.1f}%)")
```

## 5. Add BMI Status Labels

We map each numeric BMI Index to a human-readable health status.

```python
# Map index to health status
status_map = {
    0: 'Extremely Weak',
    1: 'Weak',
    2: 'Normal',
    3: 'Overweight',
    4: 'Obesity',
    5: 'Extreme Obesity'
}

df['Status'] = df['Index'].map(status_map)
print(df[['Gender', 'Height', 'Weight', 'BMI', 'Index', 'Status']].head(10))
```

## 6. BMI Distribution (Histogram)

This histogram shows how BMI values are spread across the population.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['BMI'], bins=10, color='steelblue', edgecolor='white')
plt.title('BMI Distribution Across All 500 People', fontsize=14)
plt.xlabel('BMI')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('output_bmi_histogram.jpg')
plt.show()
```

## 7. Gender & Category Distribution (Pie Charts)

Two donut charts showing:
- Left: Male vs Female split
- Right: Overall BMI category breakdown

```python
gender_counts = df['Gender'].value_counts()
category_counts = df['Status'].value_counts()

fig = plt.figure(figsize=(16, 6))
fig.suptitle('Gender Split & BMI Category Distribution', fontsize=15, fontweight='bold')

# --- Gender pie chart ---
ax1 = fig.add_subplot(121)
ax1.axis('equal')
wedges, texts, autotexts = ax1.pie(
    gender_counts,
    radius=0.8,
    explode=(0.01, 0.01),
    labels=['Female', 'Male'],
    colors=['#f7879a', '#49759c'],
    autopct='%1.1f%%',
    pctdistance=0.7,
    textprops=dict(color='k'),
    wedgeprops={'linewidth': 3, 'edgecolor': 'w'}
)
plt.setp(autotexts, size=14)
plt.setp(texts, size=13)
# Donut hole
ax1.add_artist(plt.Circle((0, 0), 0.5, color='white'))
ax1.text(0, 0, 'Gender', size=16, color='#333', ha='center', weight='bold')

# --- BMI category pie chart ---
ax2 = fig.add_subplot(122)
ax2.axis('equal')
wedges2, texts2, autotexts2 = ax2.pie(
    category_counts,
    radius=0.8,
    explode=(0.01,) * len(category_counts),
    labels=category_counts.index,
    autopct='%1.1f%%',
    pctdistance=0.5,
    textprops=dict(color='k'),
    wedgeprops={'linewidth': 3, 'edgecolor': 'w'}
)
plt.setp(autotexts2, size=14)
plt.setp(texts2, size=12)
ax2.add_artist(plt.Circle((0, 0), 0.5, color='white'))
ax2.text(0, 0, 'BMI\nStatus', size=14, color='#333', ha='center', weight='bold')

plt.tight_layout()
plt.savefig('output_pie_charts.jpg')
plt.show()
```

## 8. BMI Category Breakdown by Gender

Separate donut charts for Female and Male BMI category distributions.

```python
female_data = df[df['Gender'] == 'Female']['Status'].value_counts()
male_data   = df[df['Gender'] == 'Male']['Status'].value_counts()

fig = plt.figure(figsize=(18, 7))
fig.suptitle('BMI Category: Female vs Male', fontsize=15, fontweight='bold')

for idx, (data, label, color) in enumerate([
    (female_data, 'Female', '#f7879a'),
    (male_data,   'Male',   '#49759c')
], start=1):
    ax = fig.add_subplot(1, 2, idx)
    ax.axis('equal')
    wedges, texts, autotexts = ax.pie(
        data,
        radius=0.8,
        explode=(0.01,) * len(data),
        labels=data.index,
        autopct='%0.1f%%',
        pctdistance=0.45,
        textprops=dict(color='k'),
        wedgeprops={'linewidth': 3, 'edgecolor': 'w'}
    )
    plt.setp(autotexts, size=14)
    plt.setp(texts, size=12)
    ax.add_artist(plt.Circle((0, 0), 0.5, color='white'))
    ax.text(0, 0, label, size=18, color='#1fa774', ha='center', weight='bold')

plt.tight_layout()
plt.savefig('output_gender_bmi_breakdown.jpg')
plt.show()
```

## 9. Height vs Weight Scatter Plot

Visualizing the relationship between height and weight, colored by gender.

```python
plt.figure(figsize=(8, 6))
colors = {'Male': '#49759c', 'Female': '#f7879a'}

for gender, group in df.groupby('Gender'):
    plt.scatter(group['Height'], group['Weight'],
                label=gender, color=colors[gender], alpha=0.6, edgecolors='white', linewidth=0.5)

plt.title('Height vs Weight by Gender', fontsize=14)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.tight_layout()
plt.savefig('output_scatter.jpg')
plt.show()
```

## 10. BMI Category Count (Bar Chart)

A simple bar chart showing how many people fall into each BMI category.

```python
order = ['Extremely Weak', 'Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity']
counts = df['Status'].value_counts().reindex(order)

plt.figure(figsize=(9, 5))
bars = plt.bar(counts.index, counts.values, color='steelblue', edgecolor='white')
plt.title('Number of People per BMI Category', fontsize=14)
plt.xlabel('BMI Status')
plt.ylabel('Count')
plt.xticks(rotation=20, ha='right')

# Add value labels on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 1,
             str(int(bar.get_height())),
             ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.savefig('output_bar_chart.jpg')
plt.show()
```
